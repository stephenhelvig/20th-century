```python
import pandas, requests
from bs4 import BeautifulSoup
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import matplotlib.pyplot as plt
print("All imports OK")
```

```python
# 1) Define the URL we want to scrape
url = 'https://en.wikipedia.org/wiki/Key_events_of_the_20th_century'

# 2) Download the page
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
print("Status code:", response.status_code)

# 3) Parse the page HTML
soup = BeautifulSoup(response.text, "html.parser")

# 4) Extract all text from the page
page_text = soup.get_text(separator="\n")
print(page_text[:1000])
```

```python
# Save the scraped text to a .txt file
output_file = "20th_century_key_events.txt"

with open(output_file, "w", encoding="utf-8") as f:
    f.write(page_text)

import os
print("Saved to:", os.path.abspath(output_file))
```
