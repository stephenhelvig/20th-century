```python
import pandas as pd
```

```python
data = pd.DataFrame({"Artist": ['Swan Mack', 'Caroline Cronin', '2 Weeks Notice', 'Sharikov'],
                     "Genre": ['Indie Pop', 'Pop', 'Country', 'World'],
                     "Number of Songs": [6,8,4,75]})
```

```python
data.head()
```

```python
data['Number of Songs Updated'] = data['Number of Songs'].apply(lambda x: x+1)
```

```python
data.head()
```

```python
data['Project Size'] = data['Number of Songs'].apply(
    lambda x: 'Large' if x > 10 else ('Medium' if x >=5 else 'Small'))
```
